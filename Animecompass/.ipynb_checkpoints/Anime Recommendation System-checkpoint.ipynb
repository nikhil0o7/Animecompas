```python
def text_cleaning(text):
    text = re.sub(r'&quot;', '', text)
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub(r'.hack//', '', text)
    text = re.sub(r'&#039;', '', text)
    text = re.sub(r'A&#039;s', '', text)
    text = re.sub(r'I&#039;', 'I\'', text)
    text = re.sub(r'&amp;', 'and', text)
    text = re.sub(r'Â°', '',text)
    
    return text
```

```python
anime_df['Title'] = anime_df['Title'].apply(text_cleaning)
```

```python
anime_df.isnull().sum()
```

IMDB's weighted rating (WR) which is given as :

WR = (v/(v+m)R) + (m/(v+m)C)

where, v is the number of votes for the movie; m is the minimum votes required to be listed in the chart; R is the average rating of the movie; And C is the mean vote across the whole report



```python
C = anime_df['Rating'].mean()
C
```

```python
m = anime_df['ScoredBy'].quantile(0.85)
m
```

```python
q_animes = anime_df.copy().loc[anime_df['ScoredBy'] >= m]
q_animes.shape
```

```python
anime_df['Synopsis'] = anime_df['Synopsis'].fillna('')
```

Now we'll compute Term Frequency-Inverse Document Frequency (TF-IDF) vectors for each Synopsis.
Since we have used the TF-IDF vectorizer, calculating the dot product will directly give us the cosine similarity score. Therefore, we will use sklearn's linear_kernel() instead of cosine_similarities() since it is faster.

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(anime_df['Synopsis'])

tfidf_matrix.shape
```

```python
indices = pd.Series(anime_df.index, index=anime_df['Title']).drop_duplicates()
indices
```

```python
features = ['Genre','Producer', 'Studio']
```

Genres, Producer and Studio Based Recommender It goes without saying that the quality of our recommender would be increased with the usage of better metadata. That is exactly what we are going to do in this section. We are going to build a recommender based on the following metadata: the producer, related genres and the studio.

```python
for feature in features:
    anime_df[feature] = anime_df[feature].apply(literal_eval)
```

```python
anime_df.Type.unique()
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ","")) for i in x]
    
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ",""))
        else:
            return ""
```

```python
features = ['Genre','Producer', 'Studio', 'Type']

for feature in features:
    anime_df[feature] = anime_df[feature].apply(clean_data)
```

```python
features
```

```python
def create_sauce(x):
    return " ".join(x['Genre']) + " " + x['Type'] + " " + " ".join(x['Producer']) + " " + " ".join(x['Studio']) + " " + x['Synopsis'] + " " + " ".join(x['Studio']) 
```

```python
anime_df['sauce'] = anime_df.apply(create_sauce, axis=1)
```

```python
anime_df['sauce']
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(anime_df['sauce'])

cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
anime_df = anime_df.reset_index()
indices = pd.Series(anime_df.index, index=anime_df['Title'])
```

```python
indices
```

```python
def get_recommendations(title, cosine_sim=cosine_sim2,rating_range=(0, 10)):
    count = CountVectorizer(stop_words='english')
    count_matrix = count.fit_transform(anime_df['sauce'])

    
    idx = indices[title]
    min_rating, max_rating = rating_range

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 15 most similar movies
    sim_scores = sim_scores[1:10]
    
    movie_indices = [i[0] for i in sim_scores]
    recommended_animes = anime_df.iloc[movie_indices]

    return recommended_animes[(recommended_animes['Rating'] >= min_rating) & (recommended_animes['Rating'] <= max_rating)][['Anime_id', 'Title', 'Genre', 'Rating', 'Link']]
```

```python
anime_df['Genre_str'] = anime_df['Genre'].apply(lambda x: ' '.join(x))
```

```python
anime_df.head(5)
```

```python
tfidf_genre = TfidfVectorizer(stop_words='english')
tfidf_genre_matrix = tfidf_genre.fit_transform(anime_df['Genre_str'])

# Calculate cosine similarity based on genre vectors
cosine_sim_genre = linear_kernel(tfidf_genre_matrix, tfidf_genre_matrix)
```

```python
def get_genre_recommendations(genres, cosine_sim=cosine_sim_genre, rating_range=(0, 10)):
    # Convert the list of genres to a single string
    genre_str = ' '.join(genres)
    print(genre_str)
    tfidf_genre = TfidfVectorizer(stop_words='english')
    tfidf_genre_matrix = tfidf_genre.fit_transform(anime_df['Genre_str'])
    
    # Vectorize the passed genre string
    genre_vector = tfidf_genre.transform([genre_str])
    
    # Compute the cosine similarities
    sim_scores = linear_kernel(genre_vector, tfidf_genre_matrix).flatten()
    
    # Get sorted anime indices based on the similarity scores
    sorted_anime_indices = sim_scores.argsort()[::-1]
    
    # Filter top 15
    sorted_anime_indices = sorted_anime_indices[1:10]
    
    recommended_animes = anime_df.iloc[sorted_anime_indices]
    
    return recommended_animes[(recommended_animes['Rating'] >= rating_range[0]) & (recommended_animes['Rating'] <= rating_range[1])][['Anime_id', 'Title', 'Genre', 'Rating', 'Link']]
```
